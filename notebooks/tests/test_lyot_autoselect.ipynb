```python
%matplotlib widget
import heeps
import matplotlib.pyplot as plt
from heeps.optics import lyot_stop

conf = heeps.config.read_config(verbose=False)
conf['select_lyot'] = 'auto'
conf = heeps.config.update_config(saveconf=False, verbose=True, **conf) 
```

```python
wf = heeps.pupil.pupil(savefits=False, verbose=True, **conf)

conf['apply_ls']=False
ls = lyot_stop(wf, verbose=True, **conf)

plt.figure()
plt.imshow(heeps.util.img_processing.crop_img(np.abs(ls), 285))
```
